```python
vowels=0
consonants=0
samogloski = "aeiouyąęóAEIOUYĄĘÓ"

tekst = input('Podaj jakiś tekst: \n')
print('Podałeś : \n' + tekst)

for l in tekst:
  print(l)
  if l.isalpha():
        if l in samogloski:
            vowels += 1
        else:
            consonants += 1
print('Liczba samogłosek: ', vowels)
print('Liczba spółgłosek: ', consonants)
```
